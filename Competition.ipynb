```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
train = pd.read_csv('train.csv/train.csv')
test = pd.read_csv('test.csv/test.csv')
submission = pd.read_csv('sample_submission.csv/sample_submission.csv')

train.head()
```

```python
rowsums=train.iloc[:,2:].sum(axis=1) # tuple of comment nr and sum all values after the second column
train['clean']=(rowsums==0)          # new column in the train data frame with the data with a sum of 0 from above line
train['clean'].sum()                 # number of coulms with 0 as a result from line 1
print("Total comments = ",len(train))
print("Total clean comments = ",train['clean'].sum())
print("Total tags =",rowsums.sum())
train.head(100)
```

```python
print("Check for missing values in Train dataset")
null_check=train.isnull().sum()
print(null_check)
print("Check for missing values in Test dataset")
null_check=test.isnull().sum()
print(null_check)
print("filling NA with \"unknown\"")
train["comment_text"].fillna("unknown", inplace=True)
test["comment_text"].fillna("unknown", inplace=True)
```

```python
x=train.iloc[:,2:].sum()
print(x)
plt.figure(figsize=(15,10))
plt.bar(x.index, x.values)
plt.title("# per class")
plt.ylabel('# of Occurrences')
plt.xlabel('Type')

plt.show()
```

```python
x=rowsums.value_counts()
print(x)
#plot
plt.figure(figsize=(15,10))
plt.bar(x.index, x.values)
plt.title("Multiple tags per comment")
plt.ylabel('# of Occurrences')
plt.xlabel('# of tags')

plt.show()
```
